```python
# --------------------------------------------
# 1. Generate Individual User Report (PDF)
# --------------------------------------------
import pandas as pd
import os
from fpdf import FPDF

# Load dataset
df = pd.read_csv("../data/assigned_tasks.csv")

# Ask for user input
username = input("Enter the username (assigned_to): ").strip()

# Filter user-specific tasks
user_df = df[df["assigned_to"].str.lower() == username.lower()]

# Check if tasks exist
if user_df.empty:
    print(f"‚ö†Ô∏è No tasks found for user '{username}'")
else:
    print(f"üìã Found {len(user_df)} task(s) for user: {username}")

    # Create output folder
    os.makedirs(f"../reports/users", exist_ok=True)

    # Fonts
    FONT_PATH = r"C:\Users\Admin\ai_task_manager\fonts\DejaVuSans\ttf"

    # Initialize PDF
    pdf = FPDF()
    pdf.add_page()

    # Register Unicode font
    pdf.add_font("DejaVu", "", os.path.join(FONT_PATH, "DejaVuSans.ttf"), uni=True)
    pdf.add_font("DejaVu", "B", os.path.join(FONT_PATH, "DejaVuSans-Bold.ttf"), uni=True)

    # Title
    pdf.set_font("DejaVu", "B", 16)
    pdf.cell(200, 10, f"üìã Task Report for {username}", ln=True, align="C")

    # Summary
    pdf.set_font("DejaVu", "", 12)
    pdf.ln(10)
    pdf.cell(200, 10, f"‚úÖ Total Tasks: {len(user_df)}", ln=True)

    # Table Header
    pdf.ln(5)
    pdf.set_font("DejaVu", "B", 11)
    pdf.cell(20, 8, "ID", border=1)
    pdf.cell(80, 8, "üìù Description", border=1)
    pdf.cell(30, 8, "üìÖ Deadline", border=1)
    pdf.cell(30, 8, "‚ö° Priority", border=1)
    pdf.cell(30, 8, "üîÆ Predicted", border=1)
    pdf.ln()

    # Table Rows
    pdf.set_font("DejaVu", "", 10)
    for _, row in user_df.iterrows():
        pdf.cell(20, 8, str(row["task_id"]), border=1)
        pdf.cell(80, 8, str(row["description"])[:40], border=1)
        pdf.cell(30, 8, str(row["deadline"]), border=1)
        pdf.cell(30, 8, str(row["priority"]), border=1)
        pdf.cell(30, 8, str(row["predicted_priority"]), border=1)
        pdf.ln()

    # Save PDF
    report_path = f"../reports/users/{username}_report.pdf"
    pdf.output(report_path)

    print(f"\n‚úÖ PDF report saved at: {report_path}")
```
